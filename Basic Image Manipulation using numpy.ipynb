# Greyscale

```python
image_greyscale = np.mean(image, axis=-1)
```

```python
image_greyscale.shape
```

```python
image_greyscale[:3, :3]
```

```python
plt.imshow(image_greyscale, cmap='Greys')
```

# Black & White

```python
image_blackandwhite = np.copy(image_greyscale)
image_blackandwhite[image_blackandwhite < 128] = 0
image_blackandwhite[image_blackandwhite >= 128] = 255
```

```python
image_blackandwhite.shape
```

```python
image_blackandwhite[:3, :3]
```

```python
plt.imshow(image_blackandwhite, cmap='Greys')
```
